```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Monthly dates
datearray = pd.date_range(start='2023-01-01', periods=12, freq='M')

# Reconstructed values from the figure
processed = [160, 180, 238, 148, 180, 155, 125, 156, 126, 104, 124, 140]
received  = [None, None, None, None, 180, 160, 133, 202, 160, 139, 149, 177]

# Plot
fig, ax = plt.subplots(figsize=(10, 5))

# Blue line: Processed
ax.plot(datearray, processed, color='#2f558e', linewidth=3, marker='o', markersize=6)

# Gray line: Received
ax.plot(datearray, received, color='gray', linewidth=3, marker='o', markersize=6)

# Month labels
ax.set_xticks(datearray)
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Axis limits and style
ax.set_ylim(0, 300)
ax.set_xlim(datearray[0], datearray[-1])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(False)

# Labels for blue series
for x, y in zip(datearray[7:], processed[7:]):  # Aug-Dec
    ax.text(x, y - 12, str(y), color='#2f558e', fontsize=11,
            ha='center', va='top')

# Labels for gray series
for x, y in zip(datearray[7:], received[7:]):  # Aug-Dec
    ax.text(x, y + 10, str(y), color='gray', fontsize=11,
            ha='center', va='bottom')

# End labels
ax.text(datearray[-1] + pd.Timedelta(days=10), received[-1], 'Received',
        color='gray', fontsize=16, va='center')
ax.text(datearray[-1] + pd.Timedelta(days=10), processed[-1], 'Processed',
        color='#2f558e', fontsize=16, va='center')

plt.tight_layout()
plt.show()
```
